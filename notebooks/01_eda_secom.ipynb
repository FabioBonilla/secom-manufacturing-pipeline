```python
import pandas as pd
import numpy as np

data = pd.read_csv('../data/raw/secom.data', sep=' ', header=None)
labels = pd.read_csv('../data/raw/secom_labels.data', sep=' ', header=None, names=['label', 'timestamp'])

print("Forma de los datos (filas, columnas):", data.shape)
print("Forma de las etiquetas:", labels.shape)
```

```python
data.columns = [f'sensor_{i+1}' for i in range(data.shape[1])]
data.head()
```

```python
labels['label'].value_counts()
```

```python
nulos_por_sensor = data.isnull().sum().sort_values(ascending=False)
nulos_por_sensor.head(20)
```

```python
porcentaje_nulos = (data.isnull().sum() / len(data)) * 100

#cuantos sensores tienen mas de 50% de datos faltantes?

print("Sensores con >50% nulos", (porcentaje_nulos > 50).sum())
print("Sensores con >30% nulos", (porcentaje_nulos > 30).sum())
print("Sensores con >10% nulos", (porcentaje_nulos > 10).sum())
print("Sensores sin ningun nulo", (porcentaje_nulos == 0).sum())
```

```python
# Eliminamos sensores que tengas mas del 50% de nulos

umbral = 0.5
columnas_a_mantener = data.columns[data.isnull().mean() <= umbral]
data_limpio = data[columnas_a_mantener]

print("Columnas orignales:", data.shape[1])
print("Columnas despues de eliminar >50% nulos:",data_limpio.shape[1])

# Asignar los nulos restantes con la mediana de cada sensor

data_limpio = data_limpio.fillna(data_limpio.median())

# Verificacion final, no deberia quedar ningun nulo

print("Nulos restantes:", data_limpio.isnull().sum().sum())
```
